# Introduction
Seaborn is another library for data visualization and it is statical ploting library which is build on top of matplotlib i.e, seaborn have all the functionalities of matplotlib and also  used for advanced data visualization techniques

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
sns.barplot(x='sex' ,y='total_bill' ,data=tips) 
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill' ,data=tips,hue='smoker') #categorical  data
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue = 'sex')
```

# Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde = False,bins=30,color='r')
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='day',y='total_bill',data=tips) #comparision
```

```python
sns.pairplot(tips) #relationship
```

```python
sns.pairplot(tips,hue='day')
```

```python
tips.corr() # correlation
```

```python
sns.heatmap(tips.corr(),annot=True)
```
